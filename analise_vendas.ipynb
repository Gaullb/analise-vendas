```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dataset
df = pd.read_csv("vendas.csv")

# Mostrar primeiras linhas
df.head()
```

```python
# Criar a coluna de faturamento
df["Faturamento"] = df["Quantidade"] * df["PrecoUnitario"]

# Exibir as 5 primeiras linhas para conferir
df.head()
```

```python
# Receita total
print("💰 Receita Total:", df["Faturamento"].sum())

# Produto mais vendido
produto_top = df.groupby("Produto")["Quantidade"].sum().idxmax()
print("🏆 Produto mais vendido:", produto_top)

# Região com maior faturamento
regiao_top = df.groupby("Regiao")["Faturamento"].sum().idxmax()
print("🌍 Região com maior faturamento:", regiao_top)
```

```python
# Faturamento por região
df.groupby("Regiao")["Faturamento"].sum().plot(kind="bar", title="Faturamento por Região")
plt.ylabel("Faturamento")
plt.show()

# Evolução do faturamento ao longo do tempo
df.groupby("Data")["Faturamento"].sum().plot(kind="line", title="Evolução das Vendas")
plt.ylabel("Faturamento")
plt.xticks(rotation=45)
plt.show()

# Produtos mais vendidos
df.groupby("Produto")["Quantidade"].sum().plot(kind="barh", title="Top Produtos Vendidos")
plt.xlabel("Quantidade")
plt.show()
```
